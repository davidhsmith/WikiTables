# COVID-19 data at [The COVID Tracking Project - Homepage](https://covidtracking.com/)

The [The COVID Tracking Project - Homepage](https://covidtracking.com/) provides this data in the form of JSON and CSV files.  The data is updated daily at 

```python
import pandas as pd
```

```python
urls = ['https://covidtracking.com/api/states.csv',
        'http://covidtracking.com/api/states/daily.csv',
        # https://covidtracking.com/api/states/info.csv
        'https://covidtracking.com/api/us/daily.csv',
       ]
urls

dfs = [pd.read_csv(url, parse_dates=True) for url in urls]
```

```python
for df in dfs:
    display(df)
```

```python
dfs[0].drop(columns=['notes', 'lastUpdateEt', 'dateModified'])
dfp = dfs[0][['state', 'positive', 'negative', 'pending', 'death', 'total', 'dateChecked']]
dfp
```

Change date columns to Pandas-friendly data type.

```python
for df in [dfs[i] for i in [1,2]]:
    df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
#     display(df)
```

```python
df = dfs[2]
df['ratio'] = df['positive']/(df['positive'] + df['negative'])
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
plot1 = df.plot(x='date', y=['positive', 'negative', 'posNeg', 'total'], figsize=(20,5),
        title="COVID-19 Testing history for USA", grid='both')
plot2 = df.plot(x='date', y='ratio', grid='both', figsize=(20,5),
        title='COVID-19 positive/total ratio History for USA')
```

```python
plot1.figure.savefig('./images/tests_by_date_USA.png')
plot2.figure.savefig('./images/test_ratios_by_date_USA.png')
```

```python
population = 327
df = dfs[2]

test_per_day = df['total'].diff(periods=-1).reset_index(drop=True)/population
day = df['date'][1:]
dfp = pd.DataFrame({'day':day, 'tpd': test_per_day})[:-1]
plot3 = dfp.plot(x='day', y='tpd', grid='both', figsize=(20,5),
         title='COVID-19: Tests Per Day for USA', style='-o')

plot3.axhspan(392-10, 392+10)
plot3.annotate('South Korea at 20,000 tests/day', ('03-05-2020', 392+15))
plot3.axhspan(275-10, 275+10)
plot3.annotate('South Korea at 14,019 tests/day (3/26/2020)', ('03-05-2020', 275+15))

plot3.figure.savefig('./images/tpd_USA_South_Korea_ref.png')
```

## What about any State?

```python
population = {'NY':19.54, 'WA': 7.54, 'CA':39.56, 'OR':4.19, 'ID':1.75, 'NM':2.10, 'LA':4.66, 'IL':12.74, 'FL':39.56, 'TX':28.7}
state = 'NY'
df = dfs[1]
for state in ('WA', 'OR', 'NM', 'NY', 'LA', 'IL', 'TX', 'FL', 'IL'):
    df = dfs[1]
    select = df['state']==state
    df_select = df[select]
    # dfs.reset_index(inplace=True, drop=True)
    df_select

    test_per_day = df_select['total'].diff(periods=-1)/population[state]
    day = pd.to_datetime(df_select['date'], format='%Y%m%d')
    dfp = pd.DataFrame({'day':day, 'tpd': test_per_day})[:-1]
    plot3 = dfp.plot(x='day', y='tpd', grid='both', figsize=(20,5),
                    legend=['Test Per Day'], style='o-', 
                    title='COVID-19: Tests Per Day for {}'.format(state))
    plot3.set_xlabel('Date')
    plot3.set_ylabel('Tests per day per million')

    plot3.axhspan(392-10, 392+10)
    plot3.annotate('South Korea at 20,000 tests/day', ('03-07-2020', 392+15))
    plot3.axhspan(275-10, 275+10)
    plot3.annotate('South Korea at 14,019 tests/day (3/26/2020)', ('03-07-2020', 275+15))

    plot3.figure.savefig('./images/tpd_history_{}_SKref.png'.format(state))
```
