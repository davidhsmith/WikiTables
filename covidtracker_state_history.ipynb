```python
import pandas as pd
```

```python
urls = ['https://covidtracking.com/api/states.csv',
        'http://covidtracking.com/api/states/daily.csv',
        # https://covidtracking.com/api/states/info.csv
        'https://covidtracking.com/api/us/daily.csv',
       ]
urls
```

```python
dfs = [pd.read_csv(url) for url in urls]
```

```python
# dfs[0].drop(columns=['notes', 'lastUpdateEt', 'dateModified'])
dfs[0][['state', 'positive', 'negative', 'pending', 'death', 'total']]
```

```python
for df in dfs:
    display(df)
```

```python
df = dfs[2]
df['ratio'] = df['positive']/(df['positive'] + df['negative'])
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.plot(x='date', y=['positive', 'negative', 'posNeg', 'total'], title="COVID-19 Testing", grid='both')
df.plot(x='date', y='ratio')
```

```python
test_per_day = df['total'].diff(periods=-1).reset_index(drop=True) # .plot(grid='both')
day = df['date'][1:]
dfp = pd.DataFrame({'day':day, 'tpd': test_per_day})[1:]
dfp.plot(x='day', y='tpd', grid='both', legend=['Test Per Day'])
```
