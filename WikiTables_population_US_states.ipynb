## WikiTables_population_US_states

```python
import pandas as pd
```

```python
url  = 'https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population'
df = pd.read_html(url)[0]
df.head(5)
```

Ok, that's kind of a mess, but Pandas can make short work of it.

```python
df_pop = df[df.columns[2:4]]
df_pop.columns = ['state', 'population']
df_pop.sort_values('state')
df_pop = df_pop.sort_values('state')
df_pop.set_index('state', inplace=True)
df_pop
```

Now let's try to fill in 2-letter postal codes.

```python
url = 'https://simple.wikipedia.org/wiki/U.S._postal_abbreviations#States'
df_state_postal_codes = pd.read_html(url, skiprows=1, header=0)[0]
df_state_postal_codes.columns = ['postal_code', 'state']
df_state_postal_codes.set_index('state', inplace=True)
df_state_postal_codes
```

```python
df_state_postal_codes.at['Wyoming', 'postal_code']
```

```python
df_state_populations = df_state_postal_codes.merge(df_pop, left_index=True, right_index=True).reset_index()
df_state_populations
```

So this is looking pretty usable.  Save it in CSV for use by analysis programs.

```python
df_state_populations.to_csv('data/populations_vs_state.csv', index=False)
```
